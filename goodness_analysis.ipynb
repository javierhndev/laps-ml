```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename_240918='datasets/galadriel_dataset_24_09_18_high_res.h5'
df_input_240918=pd.read_hdf(filename_240918,'df_input')

filename_240917='datasets/galadriel_dataset_24_09_17.h5'
df_input_240917=pd.read_hdf(filename_240917,'df_input')

filename_240828='datasets/galadriel_dataset_24_08_28.h5'
df_input_240828=pd.read_hdf(filename_240828,'df_input')

filename_250313='datasets/galadriel_dataset_25_03_13_high_res.h5'
df_input_250313=pd.read_hdf(filename_250313,'df_input')
```

```python
df_input_250313
```

```python
#reference values
beta2=35000
beta3=0
beta4=-4.5e6
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
plot_3d= ax.scatter(df_input_240918['order2']-beta2, df_input_240918['order3']-beta3, df_input_240918['order4']-beta4,
                    c=df_input_240918['goodness'],
                    cmap='jet',vmin=1.0,vmax=1.7)
#ax.scatter(df_input_bad['order2']-beta2, df_input_bad['order3']-beta3, df_input_bad['order4']-beta4,c=df_input_bad['goodness'],
#                    cmap='jet',vmin=1.0,vmax=1.8,alpha=0.02)
ax.set_xlabel(r'$\delta\beta_2$',fontsize=14)
ax.set_ylabel(r'$\delta\beta_3$',fontsize=14)
ax.set_zlabel(r'$\delta\beta_4$',fontsize=14)
ax.view_init(elev=20., azim=50)
#plt.title('Goodness')
cbar = plt.colorbar(plot_3d,fraction=0.025, pad=-0.01)
cbar.ax.set_ylabel('Goodness', rotation=270,labelpad=10,fontsize=14)
ax.set_box_aspect(None, zoom=0.85)
plt.show()
#crop the figure with any online tool to remove the boundaries
```
